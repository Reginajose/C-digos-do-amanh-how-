# 3.2 Calculando as primeiras probabilidades

```python
probabilidade_3_irmaos = len(evento_3_irmaos) / total_de_alunos
print(f"A probabilidade de um aluno ter exatamente 3 irmãos é de {probabilidade_3_irmaos:.2%}.")
```

```python
prob_experimental_170_irmaos = len(evento_mais_170) / total_de_alunos
print(f"A probabilidade de um aluno ter mais de 170 cm de altura é de {prob_experimental_170_irmaos:.2%}.")
```

```python
# Simulando a Lei dos Grandes Números para sortear um aluno com 3 irmãos
print('\n --- Simulando ---')
num_sorteios = 100000
contagem_3_irmaos = 0
for _ in range(num_sorteios):
  aluno_sorteado = df_dados_turma.sample(n=1) # Sorteando uma linha do df aleatória
  if aluno_sorteado['qtd_irmaos'].iloc[0] == 3:
    contagem_3_irmaos += 1
prob_experimental_3_irmaos = contagem_3_irmaos / num_sorteios
print(f"Sorteamos um aluno {num_sorteios} vezes.")
print(f"A partir desses sorteios, em {contagem_3_irmaos} vezes o aluno possuía exatamente 3 irmãos.")
print(f"A probabilidade SIMULADA e não TEÓRICA de ter 3 irmão foi de: {prob_experimental_3_irmaos}")
```

```python
# Simulando a Lei dos Grandes Números para sortear um aluno com 170 irmãos
print('\n --- Simulando ---')
num_sorteios = 100000
contagem_170_irmaos = 0
for _ in range(num_sorteios):
  aluno_sorteado = df_dados_turma.sample(n=1) # Sorteando uma linha do df aleatória
  if aluno_sorteado['qtd_irmaos'].iloc[0] == 170:
    contagem_170_irmaos += 1
prob_experimental_170_irmaos = contagem_170_irmaos / num_sorteios
print(f"Sorteamos um aluno {num_sorteios} vezes.")
print(f"A partir desses sorteios, em {contagem_170_irmaos} vezes o aluno possuía exatamente 170 irmãos.")
print(f"A probabilidade SIMULADA e não TEÓRICA de ter 170 irmão foi de: {prob_experimental_170_irmaos}")
```
