```python
#IMPORTANDO DIRETAMENTE DA MÁQUINA
from google.colab import files
import pandas as pd
uploaded = files.upload()
df = pd.read_excel('TS_UF_20250507 (1).xlsx')
df.head()
```

```python
#EXCLUINDO AS COLUNAS SEM DADOS
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

```python
#INFORMAÇÃO SOBRE OS DADOS DO DASET
df.info()
```

```python
#SUBSTITUINDO OS NULOS POR ZERO
df = df.fillna(0)
df.head()
```

```python
#EXCLUÍNDO UMA COLUNA
df_sem_CO_UF = df.drop(columns=["CO_UF"])
df_sem_CO_UF.head()
```

```python
#RENOMEANDO AS COLUNAS
# Tudo minúsculo
df.columns = df.columns.str.lower()

# Substituir espaços por underline
df.columns = df.columns.str.replace(" ", "_")

# Exemplo: deixar minúsculo + underline
df.columns = df.columns.str.lower().str.replace(" ", "_")

df.shape
```

```python
#REMOVER DADOS DUPLICADOS
df.duplicated().sum()
print(df.duplicated())
df = df.drop_duplicates()
```

```python
df.to_excel('df_sem_CO_UF.xlsx', index=False)
```
